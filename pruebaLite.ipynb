```python
import pylab
import random

def f(x):
    return x*x
pylab.figure(1) #create figure 1
x=[]
y=[]
num=0.1

for kkk in range(100):
    num=random.random()*5+3
    x.append(num)
    y.append(f(num))
    
pylab.plot(x, y) #draw on figure 1
pylab.title(' Compounded Annually f(x)')
pylab.xlabel('Years of Compounding')
pylab.ylabel('si es $x^2$')
pylab.show()
pylab.savefig('karitouno') #save figure 2
```
